```python

import os
import pandas as pd
from PIL import Image, ImageStat
import cv2

# Chemin du répertoire principal
test_directory = '/Users/nicolasrince/Downloads/dataset/dataset/test'
train_directory = '/Users/nicolasrince/Downloads/dataset/dataset/train'

# Fonction pour extraire les informations d'une image
def get_image_info(file_path, dataset_type, species, category, status):
    image = Image.open(file_path)
    image_cv = cv2.imread(file_path)
    
    # Taille de l'image
    width, height = image.size
    
    # Poids de l'image
    file_size = os.path.getsize(file_path)
    
    # Contraste
    stat = ImageStat.Stat(image)
    contrast = stat.stddev[0]
    
    # Luminosité
    luminosity = stat.mean[0]
    
    # Niveaux de rouge, vert et bleu
    image_rgb=image.convert('RGB')
    r, g, b = image_rgb.split()
    red_mean = ImageStat.Stat(r).mean[0]
    green_mean = ImageStat.Stat(g).mean[0]
    blue_mean = ImageStat.Stat(b).mean[0]
    
    # Flou (variance de la Laplacienne)
    gray = cv2.cvtColor(image_cv, cv2.COLOR_BGR2GRAY)
    variance_of_laplacian = cv2.Laplacian(gray, cv2.CV_64F).var()
    
    return {
        'FileName': os.path.basename(file_path),
        'Extension': os.path.splitext(file_path)[1].lower(),
        'DatasetType': dataset_type,
        'Espèce': species,
        'Catégorie': category,
        'Statut': status,
        'FileSize': file_size,
        'Width': width,
        'Height': height,
        'Contrast': contrast,
        'Luminosity': luminosity,
        'RedMean': red_mean,
        'GreenMean': green_mean,
        'BlueMean': blue_mean,
        'Blur': variance_of_laplacian
    }

# Fonction pour lister les images et extraire les informations
def get_images_info(directory, dataset_type):
    subdirectories = [d for d in os.listdir(directory) if os.path.isdir(os.path.join(directory, d))]
    data = []
    for subdir in subdirectories:
        subdir_path = os.path.join(directory, subdir)
        species, category = subdir.split('___')
        status = 'Sain' if category == 'healthy' else 'Malade'
        for file in os.listdir(subdir_path):
            file_path = os.path.join(subdir_path, file)
            if os.path.isfile(file_path):
                image_info = get_image_info(file_path, dataset_type, species, category, status)
                data.append(image_info)
    return pd.DataFrame(data)

# Récupérer les informations des images pour les répertoires de test et de train
test_images_df = get_images_info(test_directory, 'test')
train_images_df = get_images_info(train_directory, 'train')

# Fusionner les DataFrames des images de test et de train
images_df = pd.concat([test_images_df, train_images_df], ignore_index=True)

# Afficher le DataFrame des images
display(images_df)
```

```python

from scipy.stats import ttest_ind

# Séparer les données en deux groupes : plantes saines et plantes malades
healthy_images = images_df[images_df['Statut'] == 'Sain']
unhealthy_images = images_df[images_df['Statut'] == 'Malade']

# Effectuer un test t de Student pour chaque couleur
colors = ['RedMean', 'GreenMean', 'BlueMean']
results = {}

for color in colors:
    t_stat, p_value = ttest_ind(healthy_images[color], unhealthy_images[color], equal_var=False)
    results[color] = {'t_stat': t_stat, 'p_value': p_value}

# Afficher les résultats
for color, result in results.items():
    print(f"Test t de Student pour {color}:")
    print(f"  Statistique t: {result['t_stat']}")
    print(f"  Valeur p: {result['p_value']}\n")
```
